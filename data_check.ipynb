```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException
from bs4 import BeautifulSoup as bs
from datetime import datetime, timedelta
import time
import pandas as pd
import re
import zipfile
import os
import requests
from tqdm import tqdm
from webdriver_manager.chrome import ChromeDriverManager

url_base = 'https://www.koreabaseball.com/Record/Player/{category}'
category_list = ['PitcherBasic/Basic1.aspx', 'PitcherBasic/Basic2.aspx']

driver = webdriver.Chrome(ChromeDriverManager().install())
driver.implicitly_wait(3)

# 팀별로 데이터를 저장할 딕셔너리
team_data_dict = {}

# 카테고리별로 컬럼 설정
columns_dict = {
    'PitcherBasic/Basic1.aspx': ['순위', '선수명', '팀명', 'ERA', 'G', 'W', 'L', 'SV', 'HLD', 'WPCT', 'IP', 'H', 'HR', 'BB', 'HBP', 'SO', 'R', 'ER', 'WHIP'],
    'PitcherBasic/Basic2.aspx': ['순위', '선수명', '팀명', 'ERA', 'CG', 'SHO', 'QS', 'BSV', 'TBF', 'NP', 'AVG', '2B', '3B', 'SAC', 'SF', 'IBB', 'WP', 'BK']
}

for category in category_list:
    url = url_base.format(category=category)
    driver.get(url)
    team_list = [x.strip() for x in driver.find_element(By.ID, 'cphContents_cphContents_cphContents_ddlTeam_ddlTeam').text.split('\n') if len(x) > 1]

    for team in tqdm(team_list):
        # 연도 선택
        driver.find_element(By.CLASS_NAME, 'select02').click()
        xpath_format = '//option[@value="2023"]'
        driver.find_element(By.XPATH, xpath_format).click()

        # 팀 선택 시에 요소를 다시 찾음
        try:
            team_dropdown = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.ID, 'cphContents_cphContents_cphContents_ddlTeam_ddlTeam')))
            team_dropdown.click()
            
            xpath_format2 = f'//option[text()="{team}"]'
            
            team_option = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, xpath_format2)))
            team_option.click()
        except StaleElementReferenceException:
            time.sleep(1)  # 잠시 대기 후 재시도
            team_dropdown = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.ID, 'cphContents_cphContents_cphContents_ddlTeam_ddlTeam')))
            team_dropdown.click()
            
            team_option = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, xpath_format2)))
            team_option.click()

        # 테이블 데이터 수집
        temp_list = [x.split(' ') for x in driver.find_element(By.TAG_NAME, 'table').text.split('\n')]
        temp_data = pd.DataFrame(columns=columns_dict[category])

        for i, temp in enumerate(temp_list[1:]):
            try:
                temp_data.loc[i] = temp
            except:
                ip_index = temp_list[0].index('IP')
                ip = temp[ip_index]
                del temp[13]
                temp[13] = ip + ' ' + temp[13]
                temp_data.loc[i] = temp

        # 팀별 데이터프레임에 합치기
        if team in team_data_dict:
            team_data_dict[team] = pd.merge(team_data_dict[team], temp_data, on=['순위', '선수명', '팀명'], how='outer')
        else:
            team_data_dict[team] = temp_data

        time.sleep(1)

driver.quit()

# 결과 확인 (예: 두산 팀 데이터)
team_data_dict['두산'].head()

# 결과를 CSV로 저장 (옵션)
for team, df in team_data_dict.items():
    df.to_csv(f'{team}_pitching_data.csv', index=False)
```

```python
def convert_IP(innings_str):
    if ' ' in innings_str:
        whole, fraction = innings_str.split(' ')
        if fraction == '1/3':
            fraction_value = 0.3
        elif fraction == '2/3':
            fraction_value = 0.6
        else:
            fraction_value = 0
        return float(whole) + fraction_value
    else:
        fraction = innings_str
        if fraction == '1/3':
            fraction_value = 0.3
        elif fraction == '2/3':
            fraction_value = 0.6
        else :
            fraction_value = fraction
        return fraction_value
```
