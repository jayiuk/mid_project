```python
df['WPCT'] = pd.to_numeric(df['WPCT'].replace("-", np.nan))
df.info()
```

```python
def convert_IP(innings_str):
    if ' ' in innings_str:
        whole, fraction = innings_str.split(' ')
        if fraction == '1/3':
            fraction_value = 1/3
        elif fraction == '2/3':
            fraction_value = 2/3
        else:
            fraction_value = 0
        return float(whole) + fraction_value
    else:
        return float(innings_str)
```

```python
df['IP'] = df['IP'].apply(convert_IP)
df.info()
```

```python
df.to_csv('pitcher_data.csv', index = False)
```
