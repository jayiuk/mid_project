```python
def del_comma(df):
    rows = []
    for row in df['연봉(만원)']:
        row = ''.join(row)
        row_n = re.sub('[^가-힣0-9]', '', row)
        row_ns = row_n.split()
        row_ns = ''.join(row_ns)
        rows.append(row_ns)
    print(rows)
    df['연봉(만원)'] = rows
    df = df.rename(columns = {'연봉(만원)' : '연봉'})
    df['연봉'] = pd.to_numeric(df['연봉'])
    return df
```
